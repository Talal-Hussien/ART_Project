# **Data Collection & Preprocessing**

```python
import requests
from bs4 import BeautifulSoup
import spacy
import pandas as pd
import string
from spacy.lang.en.stop_words import STOP_WORDS
import spacy

spacy.prefer_gpu()

# Load English NLP model
nlp = spacy.load("en_core_web_sm")
nlp.max_length = 3_000_000  # Increase processing limit

# Define animal list
animal_list = [
    "dog", "cat", "lion", "tiger", "elephant", "wolf", "rabbit", "fox", "bear", "eagle",
    "deer", "horse", "cow", "sheep", "goat", "buffalo", "camel", "giraffe", "zebra", "hippopotamus",
    "crocodile", "alligator", "otter", "beaver", "badger", "squirrel", "chipmunk", "rat", "mouse", "hamster",
    "porcupine", "mole", "hedgehog", "bat", "dolphin", "whale", "shark", "octopus", "seal", "walrus",
    "penguin", "parrot", "owl", "hawk", "falcon", "vulture", "peacock", "sparrow", "pigeon", "robin",
    "turkey", "chicken", "duck", "goose", "swan", "lizard", "snake", "frog", "toad", "newt",
    "salamander", "crab", "lobster", "shrimp", "jellyfish", "starfish", "fish", "coral", "butterfly", "moth", "bee",
    "wasp", "ant", "beetle", "dragonfly", "grasshopper", "cricket", "spider", "scorpion", "centipede", "millipede"
]


# Function to fetch text from a URL
def fetch_text(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    return soup.get_text()

# Function to extract sentences containing animal mentions
def extract_animal_sentences(text, category):
    sentences_data = []
    doc = nlp(text)
    for sent in doc.sents:
        sent_text = sent.text.strip()
        for token in sent:
            if token.text.lower() in animal_list:
                sentences_data.append({"animal name": token.text.lower(), "sentence": sent_text, "book type": category})
    return sentences_data


# Collecting all extracted data
all_data = []

for category, books_list in books.items():
    for book, url in books_list.items():
        text = fetch_text(url)
        sentences = extract_animal_sentences(text, category)
        all_data.extend(sentences)

# Convert to DataFrame
df_animals = pd.DataFrame(all_data)

```

# **Sentiment Analysis Using VADER**

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd

# Initialize sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(row):

        scores = analyzer.polarity_scores(row["sentence"])
        compound_score = scores["compound"]

        # Define sentiment categories
        if compound_score >= 0.05:
            sentiment = "positive"
        elif compound_score <= -0.05:
            sentiment = "negative"
        else:
            sentiment = "neutral"

        return pd.Series({
            "animal name": row["animal name"],
            "sentence": row["sentence"],
            "sentiment": sentiment,
            "book type": row["book type"]
        })


# Function to analyze sentiment with categorical labels
def analyze_sentiment(df):
    return df.apply(get_sentiment, axis=1)

# Apply sentiment analysis
df_sentiment = analyze_sentiment(df_animals)
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Count the number of occurrences of each sentiment per animal
animal_sentiment_counts = pd.crosstab(df_sentiment["animal name"], df_sentiment["sentiment"])

# Ensure all sentiment categories exist
for sentiment in ["positive", "negative", "neutral"]:
    if sentiment not in animal_sentiment_counts.columns:
        animal_sentiment_counts[sentiment] = 0  # Add missing columns

# Create a graph
G = nx.Graph()

# Add nodes for animals
for animal in animal_sentiment_counts.index:
    G.add_node(animal, color='lightblue')

# Add "Positive", "Negative", and "Neutral" nodes
G.add_node("Positive", color='green')
G.add_node("Negative", color='red')
G.add_node("Neutral", color='gray')

# Connect each animal to the **most frequent** sentiment
for animal, row in animal_sentiment_counts.iterrows():
    max_sentiment = row.idxmax()  # Get sentiment with highest count
    G.add_edge(animal, max_sentiment.capitalize())  # Connect to the highest sentiment

# Get node colors
node_colors = [G.nodes[n]['color'] for n in G.nodes]

# Draw the graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)  # Layout for positioning nodes
nx.draw(G, pos, with_labels=True, node_color=node_colors, edge_color='gray', node_size=800, font_size=10, font_weight="bold")

# Show the plot
plt.title("Animal Sentiment Graph")
plt.show()
```

**NSS percentage**

```python
import pandas as pd

nss_df = df_sentiment.groupby("animal name").agg(
    positive_count=("sentiment_positive", "sum"),
    negative_count=("sentiment_negative", "sum"),
    neutral_count=("sentiment_neutral", "sum")
).reset_index()

# Compute NSS percentage
nss_df["NSS percentage"] = 100 * (nss_df["positive_count"] - nss_df["negative_count"]) / (
    nss_df["positive_count"] + nss_df["negative_count"] + nss_df["neutral_count"]
)


nss_df = nss_df[["animal name", "NSS percentage"]]

# Sort the NSS DataFrame by NSS percentage in descending order
nss_df_sorted = nss_df.sort_values(by="NSS percentage", ascending=False)

nss_df_sorted[0:10]


```

# **Metaphor Bert Detection**

```python
from transformers import BertTokenizer, BertModel
import torch
import pandas as pd
from tqdm import tqdm

# Copy the dataset
df_metaphor = df_animals.copy()

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")
model.eval()  # Set to evaluation mode to disable dropout

# Move model to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Cache for already computed embeddings to avoid redundant calculations
embedding_cache = {}

def get_embedding(text):
    """Get sentence embedding using BERT with pooling."""
    if text in embedding_cache:
        return embedding_cache[text]  # Return cached embedding

    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128).to(device)

    with torch.no_grad():
        outputs = model(**inputs)

    embedding = outputs.last_hidden_state.mean(dim=1)  # Mean pooling
    embedding_cache[text] = embedding  # Cache result

    return embedding


def is_metaphor_bert(animal, sentence):
    """Compare animal embedding with sentence embedding to detect metaphor."""
    animal_embedding = get_embedding(animal)
    sentence_embedding = get_embedding(sentence)

    similarity = torch.cosine_similarity(animal_embedding, sentence_embedding).item()

    return similarity < 0.2  # Lower similarity means metaphor

# Apply function to the dataset with progress bar
tqdm.pandas()
df_metaphor["metaphor"] = df_metaphor.progress_apply(
    lambda row: is_metaphor_bert(row["animal name"], row["sentence"]), axis=1
)

df_metaphor
```

```python
df_metaphor["literally"] = ~df_metaphor["metaphor"]
df_metaphor
```

**NSS percentage**

```python
import pandas as pd

nss_df_metaphor	 = df_metaphor.groupby("animal name").agg(
    metaphor_count=("metaphor", "sum"),
    literally_count=("literally", "sum")
).reset_index()

# Compute NSS percentage
nss_df_metaphor["NSS percentage"] = 100 * (nss_df_metaphor["metaphor_count"] - nss_df_metaphor["literally_count"]) / (
    nss_df_metaphor["metaphor_count"] + nss_df_metaphor["literally_count"]
)

# Select required columns
nss_df_metaphor = nss_df_metaphor[["animal name", "NSS percentage"]]

# Sort the NSS DataFrame by NSS percentage in descending order
nss_df_metaphor_sorted = nss_df_metaphor.sort_values(by="NSS percentage", ascending=False)

nss_df_metaphor_sorted


```
